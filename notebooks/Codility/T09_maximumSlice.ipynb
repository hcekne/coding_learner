## MaxProfit

### Given a log of stock prices compute the maximum possible earning

```python
A=[0]*6
A[0] = 23171
A[1] = 21011
A[2] = 21123
A[3] = 21366
A[4] = 21013
A[5] = 21367
```

```python
# this works but is way too slow
def maxProfit(A):
    N=len(A)
    #print(A)
    # edge and minimal cases
    if N in [0,1]:
        return 0

    maxSum = 0
    intermedSum = 0

    for i in range(1,N):
        for j in range(i,N):
            #print("i is "+str(i)+" and j is "+str(j))
            intermedSum = A[j]-A[i]
            
            #print("A[j] is currently equal to "+str(A[j])+" and A[j-1] is: "+str(A[j-1]) +" and the intermedSum is "+str(intermedSum))
            if intermedSum > maxSum:
                maxSum = intermedSum

    #print(A)
    return maxSum
            
        
    
```

```python
# another approach
def maxProfitFast(A):
    N=len(A)

    # 0. edge and minimal cases
    if N in [0,1]:
        return 0

    minSeq= [-1]*N
    maxSeq= [-1]*N

    minSeq[0]=A[0]
    maxSeq[N-1]=A[N-1]

    # 1. first find the min and max at each point in the sequence
    for i in range(1,N):
        #min
        if A[i]<minSeq[i-1]:
            minSeq[i]=A[i]
        else:
            minSeq[i]=minSeq[i-1]
        #max
        if A[N-i-1]>maxSeq[N-i]:
            maxSeq[N-i-1]=A[N-i-1]
        else:
            maxSeq[N-i-1]=maxSeq[N-i]

    # 2. then iterate through the sequence
    maxDiff = 0
    
    
    for i in range(N-1):
        #print("-----new iteration--- this is the "+str(i)+" round------")
        
        #print("The current maxDiff is then "+str(maxDiff))
        
        if (maxSeq[i]-minSeq[i])>maxDiff:
            maxDiff = maxSeq[i]-minSeq[i]

        #print("Max seq is equal to "+str(maxSeq[i]))
        #print("and min seq is equal to "+str(minSeq[i]))
        
        

    #print(A)
    return maxDiff
```

## Max Slice Sum

### Find a maximum sum of a compact subsequence of array elements

```python
def maxSumSliceSlow(A):
    N=len(A)

    # 0. edge cases
    if N==1:
        return A[0]

    maxSumSliceSum = A[0]
    intermediateSum = maxSumSliceSum
    
    

    for i in range(1,N):
        #print("------------starting a new loop "+str(i)+" -----------------")
        intermediateSum = A[i-1]
        #print("intermediate sum is "+str(intermediateSum))
        if intermediateSum > maxSumSliceSum:
            maxSumSliceSum = intermediateSum
        #print("the current max slice sum at the start of the loop is "+str(maxSumSliceSum))
        for j in range(i,N):
            #print("j is now "+str(j))
            intermediateSum = A[j]+intermediateSum
            #print("the intermediate sum is: "+str(intermediateSum))
            if intermediateSum > maxSumSliceSum:
                maxSumSliceSum = intermediateSum
    
    
    return maxSumSliceSum
```

```python
# this is the solution from Chat GPT-4
def maxSliceFast(A):
    max_ending = max_slice = A[0]
    print("we initialize the variables and set max_ending and max slice equal to A[0]: "+str(A[0]))
    N = len(A)
          
    for i in range(1,N):
        print("-----------we enter into the loop iterating through A and i is: "+str(i)+" ----------")
        print("we evaluate whether or not to update max_ending by taking the maximum of "+str(A[i])+" and "+str(max_ending + A[i]))
        max_ending = max(A[i], max_ending + A[i])

        print("max ending is now updateed and equal to "+str(max_ending))
        print("we then evaluate whether the max_ending: "+str(max_ending)+" is larger than the max_slice: "+str(max_slice)+" in which case we update max slice")
        max_slice = max(max_slice, max_ending)
    return max_slice
```

This problem can be solved using Kadane's algorithm which is an efficient algorithm to solve the maximum subarray problem. This solution works by keeping track of the maximum sum ending at each position in the array (max_ending). If the maximum sum ending at the current position is negative, it doesn't make sense to continue the slice to the next position, so we start a new slice. The maximum slice sum seen so far is stored in max_slice.

The time complexity of this solution is O(N) as it only goes through the array once, and the space complexity is O(1) as it only keeps track of the maximum sums, which makes it an efficient solution for this problem.

```python
def maxDoubleSliceCorrect(A):
    N = len(A)
    forward = [0]*N
    backward = [0]*N
    
    for i in range(1, N-1):
        forward[i] = max(0, forward[i-1] + A[i])
    
    for i in range(N-2, 0, -1):
        backward[i] = max(0, backward[i+1] + A[i])
    
    max_double_slice = 0
    for i in range(1, N-1):
        max_double_slice = max(max_double_slice, forward[i-1] + backward[i+1])

    return max_double_slice
```

According to chat GPT-4:

The problem can be solved using dynamic programming with two passes of Kadane's algorithm, one forward and one backward. The forward pass finds the maximum subarray sum ending at each position, and the backward pass finds the maximum subarray sum starting from each position. We then iterate through the array to find the maximum double slice sum by combining the forward and backward sums at each position.

This solution works because at any position i, the maximum double slice sum is either the sum up to i-1 plus the sum from i+1 onwards, or it is the maximum double slice sum found so far. Thus we can find the maximum double slice sum in O(N) time.

It is important to note that the sum in forward[i] includes A[i] and the sum in backward[i] includes A[i], but the double slice sum does not include A[i]. That's why when we add forward[i-1] and backward[i+1] for the double slice sum, A[i] is not included.

Also, the range of i is from 1 to N-2 in the forward pass, backward pass and when finding the maximum double slice, because X, Y, Z have to be distinct and 0 ≤ X < Y < Z < N.
