## Triangle

Determine whether a triangle can be built from a given set of edges.

```python
def triangularChecker_slow(A):

    N = len(A)
    for p in range(0,N-2):
        for q in range(p+1,N-1):
            for r in range(p+2,N):
                if ((A[p]+A[q]) > A[r]) and ((A[q]+A[r]) > A[p]) and ((A[r]+A[p]) > A[q]):
                    return 1
    return 0
```

## Discs On a Plane

```python
## the solution from chat gpt 4
def discsOnAPlane(A):
    # Initialize the starting and ending points
    start_points = [i - A[i] for i in range(len(A))]
    end_points = [i + A[i] for i in range(len(A))]

    print(start_points)
    print(end_points)

    # Sort the points
    start_points.sort()
    end_points.sort()

    print(start_points)
    print(end_points)


    intersections = 0
    j = 0

    # Count the intersections
    for i in range(len(A)):
        print("this is the "+str(i)+ " loop")
        while j < len(A) and start_points[j] <= end_points[i]:
            print("j is now equal to "+str(j)+" and we thus increment intersections with it")
            intersections += j
            print("i is now equal to "+str(i)+" and we thus subtract intersections with it")
            intersections -= i
            print("intersections is now equal to "+str(intersections))
            j += 1
        if intersections > 10**7:
            return -1

    return intersections
```
