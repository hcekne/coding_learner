## Dominator

### Find an index of an array such that its value occurs at more than half of indices in the array

```python
### code goes here  ###
def dominator(A)
    N = len(A)

    if N==0:
        return -1
    if N==1:
        # index of the first
        return 0
    domDict = {}

    for i in range(N):
        if A[i] in domDict:
            counter = domDict[A[i]]
            domDict[A[i]] = counter +1
            if counter +1 > N/2:
                return i
        else:
            domDict[A[i]] = 1
    # if this has not returned a dominator yet.. it will not
    return -1
        


        
        
```
