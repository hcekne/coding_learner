```python
def numberOfPaths(n,m):
    # edge cases first
    if n==0 and m==0:
        print("0 and 0 so no paths, already there")
        return 0

    if n==1 and m==1:
        print("1 and 1 so no paths, already there")
        return 0
    
    if n==1 or m==1:
        print("all the points are after each other there is only one path")
        return 1

    # lets first declare a matrix of 0s which we can traverse
    matri = [[0]*n for i in range(m)]

    print(matri)
    # since you always come in from one direction, there are a maximum of 3 paths from each square but edges have 1 less
    
    numberOfPaths = 0

    for i in range(n):
        print("---------we are increasing i to"+str(i)+" -----------------")
        for j in range(m):
            print("i is now "+str(i)+", and we are setting j to"+str(j)+" -----------------")

            if i==(n-1) and j==(m-1):
                print("this is the final")
                return numberOfPaths

            elif i==0 and j==0:
                # do nothing
                print("0,0, do nothing")
            elif (i==0 and j<(m-1)):
                print("this is an edge but not a corner so 2 plus to paths")
                numberOfPaths += 2
            elif i==0 and j==(m-1):
                print("this is a corner so 1 plus to paths")
                numberOfPaths += 1
            elif i>0 and i<(n-1) and j==(m-1):
                print("this is an edge but not a corner so 2 plus to paths") 
            elif j==0 and i<(n-1):
                print("this is an edge but not a corner so 2 plus to paths")
                numberOfPaths += 2
            elif j==0 and i==n-1:
                print("this is a corner so 1 plus to paths")
                numberOfPaths += 1
            elif j>0 and j<(n-1) and i==(n-1):
                print("this is an edge but not a corner so 2 plus to paths") 
            else:
                print("inside the matrix incrementing by 3")
                numberOfPaths += 3

            matri[i][j] = 1

            print(matri)
                

            

    
    
```

Imagine you're in a maze, and you're trying to find a way from the start to the end. You can go up, down, left, or right, but you can't go through the same spot twice. 

Here's how the algorithm (or method) works:

1. **Start at the top-left corner:** This is your starting point.

2. **Mark your current spot as visited:** You don't want to go through the same spot twice, so you make a note that you've been here.

3. **Look around:** You check if you can move to a spot next to you (up, down, left, or right). If you can, you move there. If you can't (because you've already been there or it's outside the maze), you don't move.

4. **Move and repeat:** Once you've moved to a new spot, you go back to step 2. You mark your new spot as visited, look around, and move again.

5. **Found the end?** If you reach the bottom-right corner (the end of the maze), you've found a path! You count this as one path.

6. **Backtrack:** After you've found a path (or if you can't move anywhere from your current spot), you go back to the previous spot and unmark your current spot (because you're not there anymore). Then, you try a different direction from your previous spot. This is called "backtracking."

7. **Count all paths:** You keep doing this - moving, backtracking, and counting paths - until you've tried all possible ways.

8. **Done:** At the end, the number of paths you've counted is the total number of ways to get from the start to the end of the maze!

This method is like trying every possible way in a systematic manner. It's a bit like trying to find your way out of a real maze!

```python
# this is a more descriptive version of the same code...

# This function starts the process of finding all paths.
def start_finding_paths_with_backtracking(grid_width, grid_height):
    # Create a grid of the same size, filled with 'False' to represent that we haven't visited any cell yet.
    visited_cells = [[False]*grid_height for _ in range(grid_width)]
    
    # Start the depth-first search (DFS) from the top-left cell (0, 0).
    return depth_first_search_with_backtracking(0, 0, grid_width, grid_height, visited_cells)

# This function performs the depth-first search (DFS).
def depth_first_search_with_backtracking(current_x, current_y, grid_width, grid_height, visited_cells):
    # If we're outside the grid or the cell is already visited, this is not a valid path, so return 0.
    if current_x<0 or current_x>=grid_width or current_y<0 or current_y>=grid_height or visited_cells[current_x][current_y]:
        return 0
    
    # If we've reached the bottom-right cell, we've found a valid path, so return 1.
    if current_x == grid_width-1 and current_y == grid_height-1:
        return 1
    
    # Mark the current cell as visited.
    visited_cells[current_x][current_y] = True
    
    # Try moving in all four directions from the current cell, and add up the number of valid paths found.
    total_paths = depth_first_search_with_backtracking(current_x+1, current_y, grid_width, grid_height, visited_cells) + \
                  depth_first_search_with_backtracking(current_x-1, current_y, grid_width, grid_height, visited_cells) + \
                  depth_first_search_with_backtracking(current_x, current_y+1, grid_width, grid_height, visited_cells) + \
                  depth_first_search_with_backtracking(current_x, current_y-1, grid_width, grid_height, visited_cells)
    
    # After exploring all paths from the current cell, unmark it (for backtracking).
    visited_cells[current_x][current_y] = False
    
    # Return the total number of valid paths found.
    return total_paths

# Print the number of paths for a 3x3 grid.
print(start_finding_paths(3, 3))
```
